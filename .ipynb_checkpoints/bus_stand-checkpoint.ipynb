```python
import cv2
import pandas as pd
from ultralytics import YOLO
from tracker import *
import numpy as np
import csv
import time

# Load YOLO model
model = YOLO('yolov8s.pt')

# Define vehicle classes to track
vehicle_classes = ["car", "bus", "truck", "motorcycle"]

# Open 3 video files
videos = [
    cv2.VideoCapture('IMG_8979.MOV'),
    cv2.VideoCapture('IMG_8980.MOV'),
    cv2.VideoCapture('IMG_8981.MOV')
]

# Initialize trackers and counters for each video
trackers = [Tracker() for _ in range(3)]
vehicle_counts = [{"car": 0, "bus": 0, "truck": 0, "motorcycle": 0} for _ in range(3)]
crossed_red = [{} for _ in range(3)]
counter_down = [[] for _ in range(3)]

# ✅ Define custom red & blue line positions for each video
red_lines = [120, 150, 180]  # Red line Y-coordinates for videos 1, 2, 3
blue_lines = [220, 250, 280]  # Blue line Y-coordinates for videos 1, 2, 3
offset = 7  # Allow slight variation for tracking

while True:
    frames = []
    for i, cap in enumerate(videos):
        ret, frame = cap.read()
        if not ret:
            frames.append(np.zeros((360, 480, 3), dtype=np.uint8))  # Placeholder for empty video
            continue

        frame = cv2.resize(frame, (480, 360))  # Resize to fit 2x2 grid
        results = model.predict(frame)
        detections = []
        detected_vehicles = {}

        # Process YOLO detections
        if results[0].boxes:
            boxes = results[0].boxes.data.cpu().numpy()
            for row in boxes:
                x1, y1, x2, y2, _, d = row.astype(int)
                vehicle_type = model.names[d] if d < len(model.names) else "unknown"

                if vehicle_type in vehicle_classes:
                    detections.append([x1, y1, x2, y2])
                    detected_vehicles[(x1, y1, x2, y2)] = vehicle_type

        # Update tracker
        bbox_id = trackers[i].update(np.array(detections)) if detections else []

        # Count vehicles crossing the lines
        for bbox in bbox_id:
            x3, y3, x4, y4, vehicle_id = bbox
            cx, cy = (x3 + x4) // 2, (y3 + y4) // 2
            vehicle_type = detected_vehicles.get((x3, y3, x4, y4), "unknown")

            # ✅ Check for RED line crossing
            if red_lines[i] - offset < cy < red_lines[i] + offset:
                crossed_red[i][vehicle_id] = (cy, vehicle_type)

            # ✅ Check for BLUE line crossing (only if it crossed red first)
            if vehicle_id in crossed_red[i] and blue_lines[i] - offset < cy < blue_lines[i] + offset:
                _, v_type = crossed_red[i][vehicle_id]

                if vehicle_id not in counter_down[i]:
                    counter_down[i].append(vehicle_id)
                    vehicle_counts[i][v_type] += 1

        # ✅ Draw custom lines for each video
        cv2.line(frame, (50, red_lines[i]), (430, red_lines[i]), (0, 0, 255), 2)  # Red line
        cv2.putText(frame, "Red Line", (50, red_lines[i] - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

        cv2.line(frame, (50, blue_lines[i]), (430, blue_lines[i]), (255, 0, 0), 2)  # Blue line
        cv2.putText(frame, "Blue Line", (50, blue_lines[i] - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

        frames.append(frame)

    # ✅ Create stats panel (Same size as other videos)
    stats_panel = np.zeros((360, 480, 3), np.uint8)
    for idx in range(3):
        y_offset = idx * 110 + 40
        cv2.putText(stats_panel, f"Video {idx+1}:", (10, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        cv2.putText(stats_panel, f"Cars: {vehicle_counts[idx]['car']}", (10, y_offset + 30), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1)
        cv2.putText(stats_panel, f"Buses: {vehicle_counts[idx]['bus']}", (10, y_offset + 60), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1)
        cv2.putText(stats_panel, f"Trucks: {vehicle_counts[idx]['truck']}", (10, y_offset + 90), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1)
        cv2.putText(stats_panel, f"Bikes: {vehicle_counts[idx]['motorcycle']}", (10, y_offset + 120), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1)

    # ✅ Arrange videos into a 2x2 grid
    top_row = np.hstack((frames[0], frames[1]))
    bottom_row = np.hstack((frames[2], stats_panel))
    final_output = np.vstack((top_row, bottom_row))

    # ✅ Show output
    cv2.imshow("Multi-Video Vehicle Tracking (2x2 Layout)", final_output)

    if cv2.waitKey(1) & 0xFF == 27:
        break

# ✅ Release resources
for cap in videos:
    cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import pandas as pd
from ultralytics import YOLO
from tracker import*
import numpy as np
import csv
import time

model=YOLO('yolov8s.pt')

class_list = [
    "person", "bicycle", "car", "motorcycle", "airplane", "bus", "train", "truck", "boat",
    "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
    "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack",
    "umbrella", "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball",
    "kite", "baseball bat", "baseball glove", "skateboard", "surfboard", "tennis racket",
    "bottle", "wine glass", "cup", "fork", "knife", "spoon", "bowl", "banana", "apple",
    "sandwich", "orange", "broccoli", "carrot", "hot dog", "pizza", "donut", "cake",
    "chair", "couch", "potted plant", "bed", "dining table", "toilet", "TV", "laptop",
    "mouse", "remote", "keyboard", "cell phone", "microwave", "oven", "toaster", "sink",
    "refrigerator", "book", "clock", "vase", "scissors", "teddy bear", "hair drier", "toothbrush"
]

tracker=Tracker()
cap=cv2.VideoCapture('IMG_8979.MOV')
down={}
up={}

counter_down=[]
counter_up=[]

# ✅ CSV File Setup
csv_filename = "vehicles.csv"
# ✅ Create CSV file with headers if not exists
with open(csv_filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Vehicle ID", "Vehicle Type", "Timestamp", "Red Line Y", "Blue Line Y"])

# ✅ Store vehicle IDs that crossed the red line first
crossed_red = {}

# ✅ Store vehicle IDs that completed crossing both lines
counter_down = []

# ✅ Define red and blue lines

red_line_y = 335
blue_line_y = 355
offset = 7  # Allow slight variation due to tracking jitter

# ✅ Define vehicle classes to track
vehicle_classes = ["car", "bus", "truck", "motorcycle"]  
vehicle_counts = {"car": 0, "bus": 0, "truck": 0, "motorcycle": 0}  # ✅ Count of each type

while True:    
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (1020, 500))

    # YOLO Detection
    results = model.predict(frame)
    a = results[0].boxes.data
    a = a.detach().cpu().numpy()
    px = pd.DataFrame(a).astype("float")

    # ✅ Store detections
    detections = []
    detected_vehicles = {}  # ✅ Store detected vehicles and their types

    for index, row in px.iterrows():
        x1, y1, x2, y2 = int(row[0]), int(row[1]), int(row[2]), int(row[3])
        d = int(row[5])
        vehicle_type = class_list[d]  # Get detected class

        if vehicle_type in vehicle_classes:
            detections.append([x1, y1, x2, y2])
            detected_vehicles[(x1, y1, x2, y2)] = vehicle_type  # Store type with coordinates

    # ✅ Ensure 'detections' is not empty
    bbox_id = tracker.update(np.array(detections)) if detections else []

    # ✅ Process ALL detected vehicles
    for bbox in bbox_id:
        x3, y3, x4, y4, vehicle_id = bbox  # Get bounding box & ID
        cx, cy = (x3 + x4) // 2, (y3 + y4) // 2  # Compute center
        
        # ✅ Get vehicle type from stored detections
        vehicle_type = detected_vehicles.get((x3, y3, x4, y4), "unknown")

        # ✅ Check if the vehicle crosses the RED line first
        if red_line_y - offset < cy < red_line_y + offset:
            crossed_red[vehicle_id] = (cy, vehicle_type)  # Store ID & Type

        # ✅ Only consider vehicles that crossed red before reaching blue
        if vehicle_id in crossed_red and blue_line_y - offset < cy < blue_line_y + offset:
            _, v_type = crossed_red[vehicle_id]  # Get stored vehicle type

            # ✅ Draw circle at the blue line for valid vehicles
            cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)
            cv2.putText(frame, f"{v_type}-{vehicle_id}", (cx, cy), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 255), 2)

            # ✅ Add to counter only once
            if vehicle_id not in counter_down:
                counter_down.append(vehicle_id)
                vehicle_counts[v_type] += 1  # ✅ Increment count of detected type

                # ✅ Save data to CSV
                with open(csv_filename, mode="a", newline="") as file:
                    writer = csv.writer(file)
                    writer.writerow([vehicle_id, v_type, time.strftime("%Y-%m-%d %H:%M:%S"), crossed_red[vehicle_id][0], cy])

    # ✅ Draw reference lines
    cv2.line(frame, (260, 335), (724, 335), (0, 0, 255), 3)  # Red line
    cv2.putText(frame, 'red line', (260, 278), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)

    cv2.line(frame, (200, 355), (779, 355), (255, 0, 0), 3)  # Blue line
    cv2.putText(frame, 'blue line', (200, 315), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)

    # ✅ Display vehicle count
    cv2.putText(frame, f"Cars: {vehicle_counts['car']}  Buses: {vehicle_counts['bus']}  Trucks: {vehicle_counts['truck']}  Motorcycles: {vehicle_counts['motorcycle']}", 
                (60, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1, cv2.LINE_AA)    

    cv2.imshow("frames", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import pandas as pd
from ultralytics import YOLO
from tracker import *
import numpy as np
import csv
import time

# ✅ Use the smaller YOLO model (Nano version) for better CPU performance
model = YOLO('yolov8n.pt')

class_list = [
    "person", "bicycle", "car", "motorcycle", "airplane", "bus", "train", "truck", "boat",
    "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
    "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack",
    "umbrella", "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball",
    "kite", "baseball bat", "baseball glove", "skateboard", "surfboard", "tennis racket",
    "bottle", "wine glass", "cup", "fork", "knife", "spoon", "bowl", "banana", "apple",
    "sandwich", "orange", "broccoli", "carrot", "hot dog", "pizza", "donut", "cake",
    "chair", "couch", "potted plant", "bed", "dining table", "toilet", "TV", "laptop",
    "mouse", "remote", "keyboard", "cell phone", "microwave", "oven", "toaster", "sink",
    "refrigerator", "book", "clock", "vase", "scissors", "teddy bear", "hair drier", "toothbrush"
]

tracker = Tracker()
cap = cv2.VideoCapture('IMG_8979.MOV')

# ✅ CSV Setup
csv_filename = "vehicles.csv"
with open(csv_filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Vehicle ID", "Vehicle Type", "Timestamp", "Red Line Y", "Blue Line Y"])

# ✅ Initialize Counters and Tracking
crossed_red = {}
counter_down = []
vehicle_classes = ["car", "bus", "truck", "motorcycle"]
vehicle_counts = {v: 0 for v in vehicle_classes}

# ✅ Line Positions
red_line_y, blue_line_y, offset = 335, 355, 7
frame_skip = 2  # ✅ Skip every 2nd frame for better performance
frame_count = 0

# ✅ Enable OpenCV Optimizations
cv2.setUseOptimized(True)
cv2.setNumThreads(4)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    if frame_count % frame_skip != 0:  # ✅ Skip frames to improve performance
        continue

    frame = cv2.resize(frame, (640, 360))  # ✅ Reduce resolution for faster processing

    # ✅ YOLO Detection (Disable verbose for speedup)
    results = model.predict(frame, verbose=False)
    a = results[0].boxes.data.detach().cpu().numpy()
    px = pd.DataFrame(a).astype("float")

    detections = []
    detected_vehicles = {}

    for index, row in px.iterrows():
        x1, y1, x2, y2, _, d = map(int, row[:6])
        vehicle_type = class_list[d]

        if vehicle_type in vehicle_classes:
            detections.append([x1, y1, x2, y2])
            detected_vehicles[(x1, y1, x2, y2)] = vehicle_type

    bbox_id = tracker.update(np.array(detections)) if detections else []

    for bbox in bbox_id:
        x3, y3, x4, y4, vehicle_id = bbox
        cx, cy = (x3 + x4) // 2, (y3 + y4) // 2
        vehicle_type = detected_vehicles.get((x3, y3, x4, y4), "unknown")

        if red_line_y - offset < cy < red_line_y + offset:
            crossed_red[vehicle_id] = (cy, vehicle_type)

        if vehicle_id in crossed_red and blue_line_y - offset < cy < blue_line_y + offset:
            _, v_type = crossed_red[vehicle_id]

            cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)
            cv2.putText(frame, f"{v_type}-{vehicle_id}", (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)

            if vehicle_id not in counter_down:
                counter_down.append(vehicle_id)
                vehicle_counts[v_type] += 1

                with open(csv_filename, mode="a", newline="") as file:
                    writer = csv.writer(file)
                    writer.writerow([vehicle_id, v_type, time.strftime("%Y-%m-%d %H:%M:%S"), crossed_red[vehicle_id][0], cy])

    # ✅ Draw Lines
    cv2.line(frame, (260, 278), (724, 278), (0, 0, 255), 3)
    cv2.putText(frame, 'red line', (260, 278), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    cv2.line(frame, (200, 320), (779, 315), (255, 0, 0), 3)
    cv2.putText(frame, 'blue line', (200, 315), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    # ✅ Display Vehicle Count
    vehicle_text = f"Cars: {vehicle_counts['car']}  Buses: {vehicle_counts['bus']}  Trucks: {vehicle_counts['truck']}  Motorcycles: {vehicle_counts['motorcycle']}"
    cv2.putText(frame, vehicle_text, (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    cv2.imshow("Traffic Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

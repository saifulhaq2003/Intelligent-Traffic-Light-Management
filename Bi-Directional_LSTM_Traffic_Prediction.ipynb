```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import time

# Set random seeds for reproducibility
torch.manual_seed(42)
tf.random.set_seed(42)
np.random.seed(42)

# --- Data Preparation ---
def prepare_data():
    bus_stand_df = pd.read_csv("bsvehicles_summary.csv").iloc[:, :6]
    campus3_df = pd.read_csv("3vehicles_summary.csv").iloc[:, :6]
    campus12_df = pd.read_csv("12vehicles_summary.csv").iloc[:, :6]
    
    for df in [bus_stand_df, campus3_df, campus12_df]:
        df["Light Duration"] = df["Light Duration"].interpolate()
    
    scaler = MinMaxScaler()
    features = ["Cars", "Buses", "Trucks", "Motorcycles", "Light Duration"]
    bus_stand_scaled = scaler.fit_transform(bus_stand_df[features])
    
    def create_sequences(data, seq_length):
        X, y = [], []
        for i in range(len(data) - seq_length):
            X.append(data[i:i+seq_length, :-1])
            y.append(data[i+seq_length, -1])
        return np.array(X), np.array(y)
    
    seq_length = 4
    X, y = create_sequences(bus_stand_scaled, seq_length)
    return X, y, scaler, bus_stand_df

X, y, scaler, df = prepare_data()

# --- PyTorch LSTM ---
class TorchLSTM(nn.Module):
    def __init__(self, input_size=4, hidden_size=50):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
        
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

def train_pytorch(X_train, y_train):
    model = TorchLSTM()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    X_tensor = torch.tensor(X_train, dtype=torch.float32)
    y_tensor = torch.tensor(y_train, dtype=torch.float32)
    
    start_time = time.time()
    for epoch in range(100):
        optimizer.zero_grad()
        outputs = model(X_tensor)
        loss = criterion(outputs.squeeze(), y_tensor)
        loss.backward()
        optimizer.step()
    training_time = time.time() - start_time
    
    return model, training_time

# --- TensorFlow LSTM ---
def create_tf_lstm():
    return tf.keras.Sequential([
        tf.keras.layers.LSTM(50, input_shape=(X.shape[1], X.shape[2])),
        tf.keras.layers.Dense(1)
    ])

def train_tensorflow(X_train, y_train):
    model = create_tf_lstm()
    model.compile(optimizer='adam', loss='mse')
    
    start_time = time.time()
    model.fit(X_train, y_train, epochs=100, verbose=0)
    training_time = time.time() - start_time
    
    return model, training_time

# --- Training & Evaluation ---
results = []
preds_dict = {}

# PyTorch LSTM
pt_model, pt_time = train_pytorch(X, y)
with torch.no_grad():
    pt_preds = pt_model(torch.tensor(X, dtype=torch.float32)).numpy()
results.append({
    "Framework": "PyTorch",
    "Model": "LSTM",
    "Time (s)": pt_time
})
preds_dict["PyTorch LSTM"] = pt_preds

# TensorFlow LSTM
tf_model, tf_time = train_tensorflow(X, y)
tf_preds = tf_model.predict(X, verbose=0)
results.append({
    "Framework": "TensorFlow", 
    "Model": "LSTM",
    "Time (s)": tf_time
})
preds_dict["TensorFlow LSTM"] = tf_preds

# Calculate MSE and inverse transform
dummy = np.zeros((len(y), 5))
dummy[:, -1] = y
actual = scaler.inverse_transform(dummy)[:, -1]

for name, preds in preds_dict.items():
    dummy[:, -1] = preds.flatten()
    scaled_preds = scaler.inverse_transform(dummy)[:, -1]
    mse = mean_squared_error(actual, scaled_preds)
    for result in results:
        if f"{result['Framework']} {result['Model']}" == name:
            result["MSE"] = mse
            result["Predictions"] = scaled_preds

# --- Results Table (No tabulate needed) ---
print("\nPerformance Comparison:")
print("+" + "-"*40 + "+")
print("| {:<10} | {:<8} | {:<8} | {:<8} |".format(
    "Framework", "Model", "MSE", "Time(s)"))
print("+" + "-"*40 + "+")
for result in results:
    print("| {:<10} | {:<8} | {:<8.2f} | {:<8.2f} |".format(
        result["Framework"],
        result["Model"],
        result["MSE"],
        result["Time (s)"]
    ))
print("+" + "-"*40 + "+")

# --- Visualization ---
plt.figure(figsize=(14, 6))
plt.plot(df.index[seq_length:], actual, 'k-', linewidth=2, label="Actual")

colors = {'PyTorch': 'blue', 'TensorFlow': 'red'}
for result in results:
    name = f"{result['Framework']} {result['Model']}"
    plt.plot(df.index[seq_length:], result["Predictions"],
             label=f"{name} (MSE: {result['MSE']:.2f})",
             color=colors[result["Framework"]],
             linestyle='--')

plt.xlabel("Time Index", fontsize=12)
plt.ylabel("Light Duration (seconds)", fontsize=12)
plt.title("LSTM Performance: PyTorch vs. TensorFlow", fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

# Load datasets
bus_stand_df = pd.read_csv("bsvehicles_summary.csv")
campus3_df = pd.read_csv("3vehicles_summary.csv")
campus12_df = pd.read_csv("12vehicles_summary.csv")

# Remove unnecessary unnamed columns (keep only relevant ones)
bus_stand_df = bus_stand_df.iloc[:, :6]
campus3_df = campus3_df.iloc[:, :6]
campus12_df = campus12_df.iloc[:, :6]

# Fill missing 'Light Duration' using linear interpolation
for df in [bus_stand_df, campus3_df, campus12_df]:
    df["Light Duration"] = df["Light Duration"].interpolate()

# Normalize features using MinMaxScaler
scaler = MinMaxScaler()
features = ["Cars", "Buses", "Trucks", "Motorcycles", "Light Duration"]
bus_stand_scaled = scaler.fit_transform(bus_stand_df[features])
campus3_scaled = scaler.fit_transform(campus3_df[features])
campus12_scaled = scaler.fit_transform(campus12_df[features])

# Combine all datasets for training
all_data = np.vstack([bus_stand_scaled, campus3_scaled, campus12_scaled])

# Convert DataFrame to PyTorch tensors
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length, :-1])  # All features except 'Light Duration'
        y.append(data[i + seq_length, -1])  # Target: 'Light Duration'
    if len(X) == 0:
        return torch.empty(0), torch.empty(0)  # Return empty tensors if not enough data
    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

seq_length = 4
X_train, y_train = create_sequences(all_data, seq_length)
train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=8, shuffle=True) if len(X_train) > 0 else None

# Define BiLSTM Model
class BiLSTMTrafficModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(BiLSTMTrafficModel, self).__init__()
        self.bilstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, output_size)  # Multiply by 2 for bidirectional output
    
    def forward(self, x):
        lstm_out, _ = self.bilstm(x)
        return self.fc(lstm_out[:, -1, :])

# Define LSTM Model
class LSTMTrafficModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMTrafficModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        return self.fc(lstm_out[:, -1, :])

# Model Initialization
input_size = 4  # Cars, Buses, Trucks, Motorcycles
hidden_size = 50  # Updated to the best hyperparameter
num_layers = 2  # Updated to the best hyperparameter
output_size = 1

# Initialize BiLSTM and LSTM models
bilstm_model = BiLSTMTrafficModel(input_size, hidden_size, num_layers, output_size)
lstm_model = LSTMTrafficModel(input_size, hidden_size, num_layers, output_size)

# Loss and Optimizer
criterion = nn.MSELoss()
bilstm_optimizer = optim.Adam(bilstm_model.parameters(), lr=0.001)  # Updated to the best hyperparameter
lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=0.001)  # Updated to the best hyperparameter

# Training Loop for BiLSTM
if train_loader:
    epochs = 30  # Updated to the best hyperparameter
    for epoch in range(epochs):
        for X_batch, y_batch in train_loader:
            bilstm_optimizer.zero_grad()
            y_pred = bilstm_model(X_batch)
            loss = criterion(y_pred.squeeze(), y_batch)
            loss.backward()
            bilstm_optimizer.step()
        if (epoch + 1) % 10 == 0:
            print(f"BiLSTM Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

# Training Loop for LSTM
if train_loader:
    epochs = 30  # Updated to the best hyperparameter
    for epoch in range(epochs):
        for X_batch, y_batch in train_loader:
            lstm_optimizer.zero_grad()
            y_pred = lstm_model(X_batch)
            loss = criterion(y_pred.squeeze(), y_batch)
            loss.backward()
            lstm_optimizer.step()
        if (epoch + 1) % 10 == 0:
            print(f"LSTM Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

# Function to predict light duration for each timestamp
def predict_all_durations(model, data, seq_length):
    predictions = []
    model.eval()
    with torch.no_grad():
        # Initialize the sequence with the first `seq_length` timestamps
        sequence = data[:seq_length, :-1]  # Exclude 'Light Duration'
        for i in range(seq_length, len(data)):
            # Predict the next light duration
            X = torch.tensor(sequence[-seq_length:], dtype=torch.float32).unsqueeze(0)
            y_pred = model(X).item()
            predictions.append(y_pred)
            
            # Update the sequence with the new prediction
            new_row = data[i, :-1]  # Use the actual features (Cars, Buses, Trucks, Motorcycles)
            sequence = np.vstack([sequence, new_row])
    
    # Inverse transform the predictions to the original scale
    predictions = scaler.inverse_transform(np.array([[0, 0, 0, 0, pred] for pred in predictions]))[:, -1]
    return predictions

# Predict light durations for each dataset using BiLSTM
bilstm_bus_stand_predictions = predict_all_durations(bilstm_model, bus_stand_scaled, seq_length)
bilstm_campus3_predictions = predict_all_durations(bilstm_model, campus3_scaled, seq_length)
bilstm_campus12_predictions = predict_all_durations(bilstm_model, campus12_scaled, seq_length)

# Predict light durations for each dataset using LSTM
lstm_bus_stand_predictions = predict_all_durations(lstm_model, bus_stand_scaled, seq_length)
lstm_campus3_predictions = predict_all_durations(lstm_model, campus3_scaled, seq_length)
lstm_campus12_predictions = predict_all_durations(lstm_model, campus12_scaled, seq_length)

# Extract actual light durations from the datasets
actual_bus_stand_durations = bus_stand_df["Light Duration"].values[seq_length:]
actual_campus3_durations = campus3_df["Light Duration"].values[seq_length:]
actual_campus12_durations = campus12_df["Light Duration"].values[seq_length:]

# Plot the results
def plot_predictions(bilstm_predictions, lstm_predictions, actual_durations, title):
    plt.figure(figsize=(10, 6))
    plt.plot(bilstm_predictions, label="BiLSTM Predictions")
    plt.plot(lstm_predictions, label="LSTM Predictions")
    plt.plot(actual_durations, label="Actual Light Durations", linestyle="--", alpha=0.7)
    plt.xlabel("Timestamp")
    plt.ylabel("Light Duration (seconds)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Plot predictions for each dataset
plot_predictions(bilstm_bus_stand_predictions, lstm_bus_stand_predictions, actual_bus_stand_durations, "Bus Stand Predictions: BiLSTM vs LSTM vs Actual")
plot_predictions(bilstm_campus3_predictions, lstm_campus3_predictions, actual_campus3_durations, "Campus 3 Predictions: BiLSTM vs LSTM vs Actual")
plot_predictions(bilstm_campus12_predictions, lstm_campus12_predictions, actual_campus12_durations, "Campus 12 Predictions: BiLSTM vs LSTM vs Actual")
```

# Why Use BiLSTM Instead of LSTM?

### 1. **Bidirectional Context**
- BiLSTM processes sequences in **both forward and backward directions**, capturing dependencies from past and future timesteps.
- Useful for tasks like traffic light prediction, where future traffic conditions (e.g., approaching vehicles) influence light duration.

### 2. **Improved Performance**
- BiLSTM learns **more complex patterns** by combining forward and backward passes, making it better suited for sequential data with intricate dependencies.

### 3. **Long-Term Dependencies**
- BiLSTM handles **long-term dependencies** more effectively than LSTM by leveraging bidirectional context, crucial for modeling trends like rush hour traffic.

### 4. **Richer Feature Representation**
- BiLSTM combines forward and backward representations, providing a **more comprehensive understanding** of the data.

### 5. **Empirical Evidence**
- BiLSTM often **outperforms LSTM** in sequence modeling tasks, as it better captures bidirectional relationships.

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import GridSearchCV, KFold, ParameterGrid
from sklearn.base import BaseEstimator, RegressorMixin
import matplotlib.pyplot as plt

# Load datasets
bus_stand_df = pd.read_csv("bsvehicles_summary.csv")
campus3_df = pd.read_csv("3vehicles_summary.csv")
campus12_df = pd.read_csv("12vehicles_summary.csv")

# Remove unnecessary unnamed columns (keep only relevant ones)
bus_stand_df = bus_stand_df.iloc[:, :6]
campus3_df = campus3_df.iloc[:, :6]
campus12_df = campus12_df.iloc[:, :6]

# Fill missing 'Light Duration' using linear interpolation
for df in [bus_stand_df, campus3_df, campus12_df]:
    df["Light Duration"] = df["Light Duration"].interpolate()

# Normalize features using MinMaxScaler
scaler = MinMaxScaler()
features = ["Cars", "Buses", "Trucks", "Motorcycles", "Light Duration"]
bus_stand_scaled = scaler.fit_transform(bus_stand_df[features])
campus3_scaled = scaler.fit_transform(campus3_df[features])
campus12_scaled = scaler.fit_transform(campus12_df[features])

# Combine all datasets for training
all_data = np.vstack([bus_stand_scaled, campus3_scaled, campus12_scaled])

# Convert DataFrame to PyTorch tensors
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length, :-1])  # All features except 'Light Duration'
        y.append(data[i + seq_length, -1])  # Target: 'Light Duration'
    if len(X) == 0:
        return torch.empty(0), torch.empty(0)  # Return empty tensors if not enough data
    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

seq_length = 4
X, y = create_sequences(all_data, seq_length)

# Define BiLSTM Model
class BiLSTMTrafficModel(nn.Module):
    def __init__(self, hidden_size=50, num_layers=2):
        super(BiLSTMTrafficModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.bilstm = nn.LSTM(input_size=4, hidden_size=hidden_size, num_layers=num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, 1)  # Multiply by 2 for bidirectional output
    
    def forward(self, x):
        lstm_out, _ = self.bilstm(x)
        return self.fc(lstm_out[:, -1, :])

# Wrapper class to make PyTorch model compatible with scikit-learn
class PyTorchRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, hidden_size=50, num_layers=2, epochs=50, lr=0.001):
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.epochs = epochs
        self.lr = lr
        self.model = BiLSTMTrafficModel(hidden_size, num_layers)
        self.criterion = nn.MSELoss()
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)
    
    def fit(self, X, y):
        X_tensor = torch.tensor(X, dtype=torch.float32)
        y_tensor = torch.tensor(y, dtype=torch.float32)
        train_loader = DataLoader(TensorDataset(X_tensor, y_tensor), batch_size=8, shuffle=True)
        
        self.model.train()
        for epoch in range(self.epochs):
            for X_batch, y_batch in train_loader:
                self.optimizer.zero_grad()
                y_pred = self.model(X_batch)
                loss = self.criterion(y_pred.squeeze(), y_batch)
                loss.backward()
                self.optimizer.step()
        return self
    
    def predict(self, X):
        X_tensor = torch.tensor(X, dtype=torch.float32)
        self.model.eval()
        with torch.no_grad():
            y_pred = self.model(X_tensor)
        return y_pred.squeeze().numpy()

# Define the hyperparameter grid
param_grid = {
    "hidden_size": [32, 50, 64],
    "num_layers": [1, 2, 3],
    "epochs": [30, 50, 100],
    "lr": [0.001, 0.01]
}

# Initialize the PyTorch regressor
pytorch_regressor = PyTorchRegressor()

# Perform grid search with cross-validation using GridSearchCV
grid_search = GridSearchCV(
    estimator=pytorch_regressor,
    param_grid=param_grid,
    cv=KFold(n_splits=5, shuffle=True, random_state=42),
    scoring="neg_mean_squared_error",  # Use negative MSE for scoring
    n_jobs=-1  # Use all available CPU cores
)

# Fit the grid search
grid_search.fit(X.numpy(), y.numpy())

# Print the best hyperparameters and score from GridSearchCV
print("GridSearchCV Results:")
print(f"Best hyperparameters: {grid_search.best_params_}")
print(f"Best score (negative MSE): {grid_search.best_score_}")

# Manual cross-validation for comparison
best_loss = float("inf")
best_params = None
manual_results = []

for params in ParameterGrid(param_grid):
    print(f"Testing hyperparameters: {params}")
    
    # K-Fold Cross-Validation
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    avg_val_loss = 0
    
    for train_idx, val_idx in kfold.split(X):
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        
        train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=8, shuffle=True)
        val_loader = DataLoader(TensorDataset(X_val, y_val), batch_size=8, shuffle=False)
        
        # Initialize the model with current hyperparameters
        model = BiLSTMTrafficModel(
            hidden_size=params["hidden_size"],
            num_layers=params["num_layers"]
        )
        
        # Define loss and optimizer
        criterion = nn.MSELoss()
        optimizer = optim.Adam(model.parameters(), lr=params["lr"])
        
        # Train and evaluate the model
        model.train()
        for epoch in range(params["epochs"]):
            for X_batch, y_batch in train_loader:
                optimizer.zero_grad()
                y_pred = model(X_batch)
                loss = criterion(y_pred.squeeze(), y_batch)
                loss.backward()
                optimizer.step()
        
        # Evaluate on the validation set
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                y_pred = model(X_batch)
                val_loss += criterion(y_pred.squeeze(), y_batch).item()
        avg_val_loss += val_loss / len(val_loader)
    
    avg_val_loss /= 5  # Average validation loss across all folds
    manual_results.append((params, avg_val_loss))
    
    # Update best hyperparameters if the current model performs better
    if avg_val_loss < best_loss:
        best_loss = avg_val_loss
        best_params = params

print("\nManual Cross-Validation Results:")
print(f"Best hyperparameters: {best_params}")
print(f"Best average validation loss: {best_loss}")

# Plot the results
grid_search_results = grid_search.cv_results_["mean_test_score"]
manual_losses = [result[1] for result in manual_results]

plt.figure(figsize=(10, 6))
plt.plot(grid_search_results, label="GridSearchCV (Negative MSE)")
plt.plot(manual_losses, label="Manual Cross-Validation (MSE)")
plt.xlabel("Hyperparameter Combination")
plt.ylabel("Score / Loss")
plt.title("Comparison of GridSearchCV and Manual Cross-Validation")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

# Load datasets
bus_stand_df = pd.read_csv("bsvehicles_summary.csv")
campus3_df = pd.read_csv("3vehicles_summary.csv")
campus12_df = pd.read_csv("12vehicles_summary.csv")

# Remove unnecessary unnamed columns (keep only relevant ones)
bus_stand_df = bus_stand_df.iloc[:, :6]
campus3_df = campus3_df.iloc[:, :6]
campus12_df = campus12_df.iloc[:, :6]

# Fill missing 'Light Duration' using linear interpolation
for df in [bus_stand_df, campus3_df, campus12_df]:
    df["Light Duration"] = df["Light Duration"].interpolate()

# Normalize features using MinMaxScaler
scaler = MinMaxScaler()
features = ["Cars", "Buses", "Trucks", "Motorcycles", "Light Duration"]
bus_stand_scaled = scaler.fit_transform(bus_stand_df[features])
campus3_scaled = scaler.fit_transform(campus3_df[features])
campus12_scaled = scaler.fit_transform(campus12_df[features])

# Combine all datasets for training
all_data = np.vstack([bus_stand_scaled, campus3_scaled, campus12_scaled])

# Convert DataFrame to PyTorch tensors
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length, :-1])  # All features except 'Light Duration'
        y.append(data[i + seq_length, -1])  # Target: 'Light Duration'
    if len(X) == 0:
        return torch.empty(0), torch.empty(0)  # Return empty tensors if not enough data
    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

seq_length = 4
X_train, y_train = create_sequences(all_data, seq_length)
train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=8, shuffle=True) if len(X_train) > 0 else None

# Define BiLSTM Model
class BiLSTMTrafficModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(BiLSTMTrafficModel, self).__init__()
        self.bilstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, output_size)  # Multiply by 2 for bidirectional output
    
    def forward(self, x):
        lstm_out, _ = self.bilstm(x)
        return self.fc(lstm_out[:, -1, :])

# Model Initialization
input_size = 4  # Cars, Buses, Trucks, Motorcycles
hidden_size = 50  # Updated to the best hyperparameter
num_layers = 2  # Updated to the best hyperparameter
output_size = 1
model = BiLSTMTrafficModel(input_size, hidden_size, num_layers, output_size)

# Loss and Optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Updated to the best hyperparameter

# Training Loop
if train_loader:
    epochs = 30  # Updated to the best hyperparameter
    for epoch in range(epochs):
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred.squeeze(), y_batch)
            loss.backward()
            optimizer.step()
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

# Function to predict light duration for each timestamp
def predict_all_durations(model, data, seq_length):
    predictions = []
    model.eval()
    with torch.no_grad():
        # Initialize the sequence with the first `seq_length` timestamps
        sequence = data[:seq_length, :-1]  # Exclude 'Light Duration'
        for i in range(seq_length, len(data)):
            # Predict the next light duration
            X = torch.tensor(sequence[-seq_length:], dtype=torch.float32).unsqueeze(0)
            y_pred = model(X).item()
            predictions.append(y_pred)
            
            # Update the sequence with the new prediction
            new_row = data[i, :-1]  # Use the actual features (Cars, Buses, Trucks, Motorcycles)
            sequence = np.vstack([sequence, new_row])
    
    # Inverse transform the predictions to the original scale
    predictions = scaler.inverse_transform(np.array([[0, 0, 0, 0, pred] for pred in predictions]))[:, -1]
    return predictions

# Predict light durations for each dataset
bus_stand_predictions = predict_all_durations(model, bus_stand_scaled, seq_length)
campus3_predictions = predict_all_durations(model, campus3_scaled, seq_length)
campus12_predictions = predict_all_durations(model, campus12_scaled, seq_length)

# Add predictions to the original dataframes
bus_stand_df["Predicted Light Duration"] = np.nan
bus_stand_df.loc[seq_length:, "Predicted Light Duration"] = bus_stand_predictions

campus3_df["Predicted Light Duration"] = np.nan
campus3_df.loc[seq_length:, "Predicted Light Duration"] = campus3_predictions

campus12_df["Predicted Light Duration"] = np.nan
campus12_df.loc[seq_length:, "Predicted Light Duration"] = campus12_predictions

# Save the results to new CSV files
bus_stand_df.to_csv("bus_stand_predictions.csv", index=False)
campus3_df.to_csv("campus3_predictions.csv", index=False)
campus12_df.to_csv("campus12_predictions.csv", index=False)

print("Predictions saved to CSV files!")
```
